```python
def avg_grd(grades):

  sum_grades = sum(grades)


  avarage = sum_grades / len(grades)
  return(avarage)


```

```python
grades_list = [55,88,29,65]
result = avg_grd(grades_list)

print(f"Avarage grade is {result:.2f}")
```

```python
def get_letter_grade(average):
    if 90 <= average <= 100:
        return "A"
    elif 80 <= average < 90:
        return "B"
    elif 70 <= average < 80:
        return "C"
    elif 60 <= average < 70:
        return "D"
    else:
        return "F"
```

```python
std_marks = [87, 42, 71, 56]
average = avg_grd(std_marks)

letter_grade = get_letter_grade(average)

print(f"The average grade is: {average:.2f}")
print(f"The corresponding letter grade is: {letter_grade}")
```
